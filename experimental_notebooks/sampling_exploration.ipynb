## Sampling Exploration

```python
from pathlib import Path
import csv
import seaborn as sns; sns.set_theme()
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import functools

pd.options.display.float_format = '{:.2f}'.format
```

```python
from transformers import AutoConfig, AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset, load_from_disk
import torch
```

```python
def param_default():
    return {
        'current_dataset' : 'DC',
        'datasets' : {
            'SG_BD' : 'code_completion_random_cut_5k_30_512_tokens', 
            'DC_SG_BD' : 'code_completion_docstring_random_cut_3.8k_30_150_tokens', 
            'DC_SG' : 'code_completion_docstring_signature_3.8k_30_150_tokens', 
            'DC': 'code_completion_docstring_5k_30_150_tokens'
        },
        'num_experiments' : 30,
        'sampling_results': '/workspaces/code-rationales/data/sampling/gpt',
        'decoded_sampling_results': '/workspaces/code-rationales/data/sampling/gpt/decoded',
        'model_name': '/workspaces/code-rationales/data/codeparrot-small/checkpoints/checkpoint-29000', 
        'cache_dir': '/workspaces/code-rationales/datax/df_cache_dir', 
    }
    
params = param_default()
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
```

### Tokenizer Loading

```python
tokenizer = AutoTokenizer.from_pretrained(param_default()['model_name'])
```

### Exploration

```python
df_sampled_input = pd.read_csv( params['sampling_results'] + '/' + params['datasets'][params['current_dataset']] +'.csv', index_col=0)
df_sampled_input_decoded = df_sampled_input.copy()
```

```python
for experiment in range(0,params['num_experiments']):
    df_sampled_input_decoded[str(experiment)] = df_sampled_input_decoded[str(experiment)].apply(lambda ids: tokenizer.decode(eval(ids)))
```

```python
df_sampled_input_decoded.to_csv(params['decoded_sampling_results'] + '/' + params['datasets'][params['current_dataset']] +'.csv')
```

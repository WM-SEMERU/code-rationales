## Token distribution graph

```python
tkn_taxonomy  = pl_taxonomy_python()
```

```python
base_path= "/workspaces/code-rationales/datax/evaluation/"
```

```python

#Load dataframe from previous evaluation with ground truth and sampling .i.e form Levenshtein analysis
lev_df = pd.read_parquet(base_path+'code_completion_docstring_5k_30_150_tokens.parquet')
```

```python
lev_df.head()
```

```python
treatments = ["outcome_0","outocome_1"]
```

```python
def count_by_taxonomy( dct_control, k_token, splits, stringent = False, flag = False ):
    "Counting the token across BPEs"
    dict_most_k_tok = dict.fromkeys( k_token, 0 )
    for k in splits:
        for k_tok in k_token:
            if stringent:
                if k_tok == k:
                    dict_most_k_tok[k_tok] +=  dct_control[ k ]
                    if flag:
                        print( k +" : "+ str(dct_control[ k ]) )
            else:
                tmp = 'Ġ'+ k_tok
                #if k_tok in k:
                if k_tok == k or tmp == k:    
                    dict_most_k_tok[k_tok] +=  dct_control[ k ]
                    if flag:
                        print( k +" : "+ str(dct_control[ k ]) )
                
    return Counter(dict_most_k_tok)
```

```python
def print_token_distribution( dict_most, ax, log = False, label=['Control'], width=0.2, title='' ):
    multiplier = 0
    keywords = dict_most[0].keys()
    x = np.arange(len(keywords))  
    for count, label_ in enumerate(label):
        offset = width * multiplier
        rects = ax.bar(x + offset, dict_most[count].values(), width=width, align='center', alpha=.6, label=label_)
        #ax.bar_label(rects, padding=3)
        multiplier +=1
    ax.set_ylabel('N. Tokens')
    ax.set_title('Token by taxonomy '+ title)
    ax.set_xticks(x + width, keywords)
    ax.legend(loc='upper left', ncols=3)

    

    if log:
        ax.set_yscale("log")
```

```python
def count_by_taxonomy_list(list_dct_control, k_token, splits, stringent = False ):
    ''''Flatting function for Source/Target datasets'''
    return [ dict( count_by_taxonomy( dct_control, k_token , splits, stringent ) ) for dct_control in list_dct_control]
```

```python
def taxonomy_counting_v1( tkn_taxonomy, dct_control, splits, stringent=False ):
    ''''Hardcoded version of taxonomy counting'''
    dict_tokens = {}
    for split in splits:
        to_flat_tax_dict = { tax :  count_by_taxonomy_list( dct_control, tkn_taxonomy[tax].values(), splits, stringent=stringent )  for tax in tkn_taxonomy }
        to_flat_tax_dict = { tax : [ functools.reduce(lambda a, b: a+b, list_dict.values() ) for list_dict in to_flat_tax_dict[tax] ] for tax in to_flat_tax_dict }
        dict_tokens[split] = [ 
                { tax: to_flat_tax_dict[tax][0] for tax in to_flat_tax_dict } #Hardcoded Output
            ]
    return dict_tokens
```

```python
focused_tokens_non = taxonomy_counting_v1(tkn_taxonomy, lev_df, treatments, stringent=False)
```

```python
data_to_print = []
for key in treatments:
    data_to_print.append(focused_tokens_non[key][0])
fig, axs = plt.subplots( figsize=(20,4), ncols=1 )
print_token_distribution( data_to_print, ax=axs,  label=treatments, log=True )
axs.legend(treatments)
plt.show()
```
